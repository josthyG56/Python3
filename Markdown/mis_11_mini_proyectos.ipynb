# Ejercicio #2 "Adivina el Número"

Este notebook proporciona un análisis del juego "Adivina el Número" implementado en Python y documenta su funcionamiento.

### Código Python


````python
import random

def jugar_adivina_el_numero():
    """
    Esta función implementa el juego "Adivina el Número".
    El jugador debe adivinar un número secreto entre 1 y 50.
    """
    numero_secreto = random.randint(1, 50)
    intentos = 0
    
    while True:
        intento = input("Adivina el número secreto entre 1 y 50: ")

        try:
            intento = int(intento)
            if intento < 1 or intento > 50:
                print("Error: Por favor, elige un número dentro del rango correcto (1-50).")
                continue
            
            intentos += 1
            if intento < numero_secreto:
                print("Demasiado bajo. Intenta de nuevo.")
            elif intento > numero_secreto:
                print("Demasiado alto. Intenta de nuevo.")
            else:
                print(f"Felicidades, ¡adivinaste el número secreto {numero_secreto} en {intentos} intentos!")
                break

            continuar = input("¿Quieres seguir jugando? (sí/no): ").lower()
            if continuar != 'si':
                print("¡Gracias por jugar!")
                break
        except ValueError:
            print("Error: Por favor, ingresa un número válido.")

def main():
    print("Bienvenido al juego Adivina el Número!")
    jugar_adivina_el_numero()

if __name__ == "__main__":
    main()
````

## Análisis del Código

El juego "Adivina el Número" tiene las siguientes características:

- La función `jugar_adivina_el_numero()` genera un número secreto aleatorio entre 1 y 50 y solicita al usuario que adivine ese número.
- El jugador tiene un número limitado de intentos para adivinar el número secreto.
- Después de cada intento, se proporciona una pista para indicar si el número es demasiado bajo o demasiado alto.
- Una vez que el jugador adivina correctamente el número secreto, se muestra un mensaje de felicitaciones junto con el número de intentos realizados.
- Después de cada juego, se le pregunta al jugador si desea continuar jugando.
- La función `main()` inicia el juego y da la bienvenida al jugador.

El código está diseñado para manejar errores como la entrada de valores no válidos y garantizar que el jugador elija un número dentro del rango especificado. El juego proporciona una experiencia interactiva y divertida para el usuario.


# Ejercicio #3 "Biografía"

En este ejercicio, crearemos un programa en Python llamado "Biografía" que solicitará información personal al usuario, como nombre, fecha de nacimiento, dirección y metas personales. El programa validará cada entrada de información para asegurarse de que esté en el formato correcto antes de mostrar un resumen de la información ingresada.

El programa utilizará expresiones regulares para validar cada tipo de información introducida por el usuario, como el nombre, la fecha de nacimiento y la dirección. Si la información ingresada no cumple con los requisitos, se solicitará al usuario que la ingrese nuevamente.

Una vez que se haya ingresado toda la información válida, el programa imprimirá un resumen de la información personal del usuario.

### Código Python 



````python
import re

def validar_nombre(nombre):
    if re.match("^[A-Za-z ]+$", nombre):
        return True
    else:
        return False

def validar_fecha(fecha):
    if re.match("^[A-Za-z]+ [0-9]{1,2}, [0-9]{4}$", fecha):
        return True
    else:
        return False

def validar_direccion(direccion):
    if re.match("^[A-Za-z0-9, ]+$", direccion):
        return True
    else:
        return False

def solicitar_informacion():
    nombre = input("¿Cuál es tu nombre? ")
    while not validar_nombre(nombre):
        print("Por favor, ingresa un nombre válido.")
        nombre = input("¿Cuál es tu nombre? ")

    fecha_nacimiento = input("¿Cuál es tu fecha de nacimiento? (Ejemplo: Jun 1, 1954) ")
    while not validar_fecha(fecha_nacimiento):
        print("Por favor, ingresa una fecha de nacimiento válida en el formato correcto.")
        fecha_nacimiento = input("¿Cuál es tu fecha de nacimiento? (Ejemplo: Jan 1, 1954) ")

    direccion = input("¿Cuál es tu dirección? ")
    while not validar_direccion(direccion):
        print("Por favor, ingresa una dirección válida.")
        direccion = input("¿Cuál es tu dirección? ")

    metas_personales = input("¿Cuáles son tus metas personales? ")

    return nombre, fecha_nacimiento, direccion, metas_personales

def imprimir_resumen(nombre, fecha_nacimiento, direccion, metas_personales):
    print("\nResumen de tu información personal:")
    print("- Nombre:", nombre)
    print("- Fecha de nacimiento:", fecha_nacimiento)
    print("- Direccion:", direccion)
    print("- Metas:", metas_personales)

def main():
    nombre, fecha_nacimiento, direccion, metas_personales = solicitar_informacion()
    imprimir_resumen(nombre, fecha_nacimiento, direccion, metas_personales)

if __name__ == "__main__":
    main()
````

# Análisis del Código

El script consta de las siguientes partes:

1. **Funciones de Validación:**
   - `validar_nombre(nombre)`: Utiliza una expresión regular para validar que el nombre ingresado contenga solo letras y espacios.
   - `validar_fecha(fecha)`: Utiliza una expresión regular para validar que la fecha de nacimiento ingresada esté en el formato correcto (por ejemplo, "Jun 1, 1954").
   - `validar_direccion(direccion)`: Utiliza una expresión regular para validar que la dirección ingresada contenga solo letras, números, comas y espacios.

2. **Función `solicitar_informacion()`:**
   - Solicita al usuario que ingrese su nombre, fecha de nacimiento y dirección.
   - Utiliza las funciones de validación para asegurarse de que cada entrada sea válida.
   - Devuelve la información ingresada por el usuario.

3. **Función `imprimir_resumen()`:**
   - Toma los datos personales ingresados por el usuario y los imprime en un formato de resumen.

4. **Función `main()`:**
   - Llama a la función `solicitar_informacion()` para obtener la información del usuario.
   - Llama a la función `imprimir_resumen()` para mostrar un resumen de la información ingresada.

5. **Bloque de Ejecución Principal:**
   - Verifica si el script se está ejecutando directamente y, si es así, llama a la función `main()`.

El código está diseñado para ser interactivo y amigable para el usuario, proporcionando retroalimentación clara sobre cualquier entrada incorrecta y mostrando un resumen ordenado de la información ingresada.

# Ejercicio #4 "CorreoSegmentador"

En este ejercicio, desarrollaremos un programa llamado "CorreoSegmentador" en Python que permitirá a los usuarios ingresar su dirección de correo electrónico. El programa analizará la dirección de correo electrónico ingresada y la segmentará en el nombre de usuario y el dominio.

El programa verificará si el dominio del correo electrónico pertenece a una lista predefinida de dominios populares como Gmail, Yahoo y Hotmail, o si es un dominio personalizado. Dependiendo del resultado de la verificación del dominio, el programa mostrará un mensaje diferente al usuario.

### Código Python 

````python
def segmentador_correo(correo):
    nombre_usuario, dominio = correo.split('@')[0], correo.split('@')[1]
    return nombre_usuario, dominio

def verificar_dominio_personalizado(dominio):
    dominios_populares = ["gmail.com", "yahoo.com", "hotmail.com"]
    if dominio.lower() in dominios_populares:
        return False
    else:
        return True

def mensaje_final(nombre_usuario, dominio):
    if verificar_dominio_personalizado(dominio):
        return f"Hey {nombre_usuario.capitalize()}, parece que tienes tu propia configuración personalizada en {dominio.capitalize()}. ¡Impresionante!"
    else:
        return f"Hey {nombre_usuario.capitalize()}, veo que tu correo está registrado con {dominio.capitalize()}. ¡Eso es genial!"

def main():
    correo = input("Ingrese su dirección de correo electrónico: ")
    nombre_usuario, dominio = segmentador_correo(correo)
    mensaje = mensaje_final(nombre_usuario, dominio)
    print(mensaje)

if __name__ == "__main__":
    main()
````

## Análisis del Código

El script consta de las siguientes partes:

1. **Función `segmentador_correo(correo)`:**
   - Esta función toma una dirección de correo electrónico como entrada y utiliza el método `split('@')` para dividir la dirección en el nombre de usuario y el dominio.
   - Retorna el nombre de usuario y el dominio.

2. **Función `verificar_dominio_personalizado(dominio)`:**
   - Esta función toma el dominio como entrada y verifica si es un dominio personalizado o uno de los dominios populares (Gmail, Yahoo, Hotmail).
   - Retorna `True` si es un dominio personalizado, y `False` si es uno de los dominios populares.

3. **Función `mensaje_final(nombre_usuario, dominio)`:**
   - Esta función toma el nombre de usuario y el dominio como entrada.
   - Utiliza la función `verificar_dominio_personalizado()` para determinar el tipo de dominio.
   - Retorna un mensaje personalizado dependiendo del resultado de la verificación del dominio.

4. **Función `main()`:**
   - Solicita al usuario que ingrese su dirección de correo electrónico.
   - Utiliza la función `segmentador_correo()` para obtener el nombre de usuario y el dominio.
   - Llama a la función `mensaje_final()` para generar un mensaje final.
   - Imprime el mensaje final en la consola.

5. **Bloque de Ejecución Principal:**
   - Verifica si el script se está ejecutando directamente y, si es así, llama a la función `main()`.

El código está diseñado para proporcionar una interacción simple pero efectiva con el usuario, segmentando su dirección de correo electrónico y proporcionando un mensaje personalizado dependiendo del dominio.


# Ejercicio #5 "LetrasAleatorias"

En este ejercicio, desarrollaremos un programa en Python llamado "LetrasAleatorias" que permitirá a los usuarios seleccionar una canción de una lista de canciones populares de diferentes artistas. El programa mostrará un menú de canciones disponibles y permitirá al usuario elegir una canción ingresando su número correspondiente.

Una vez que se haya seleccionado una canción, el programa obtendrá la letra de la canción de una API de letras de canciones en línea y la mostrará al usuario. Si el usuario desea elegir otra canción, se le dará la opción de hacerlo.

El programa utiliza la biblioteca `requests` para realizar solicitudes HTTP a la API de letras de canciones y la API `api.lyrics.ovh` para obtener la letra de las canciones seleccionadas.

### Código Python 


````python
import requests;

canciones = {
    "Justin Bieber": ["Baby", "Sorry", "Love Yourself"],
    "Drake": ["Hotline Bling", "God's Plan", "In My Feelings"],
    "Beyonce": ["Flawless", "Single Ladies", "Halo"],
    "Eminem": ["Lose Yourself", "Stan", "Love the Way You Lie"]
}

def mostrar_menu():
    print("Bienvenido, por favor elija una canción de estas 10 canciones:\n")
    numero = 1
    for artista, lista_canciones in canciones.items():
        for cancion in lista_canciones:
            print(f"{numero}. {cancion} de {artista}")
            numero += 1

def seleccionar_cancion():
    seleccion = input("\nPor favor, ingrese el número de la canción que elija (o * para elegir nuevamente): ")
    if seleccion == "*":
        mostrar_menu()
        seleccionar_cancion()
    elif seleccion.isdigit():
        seleccion = int(seleccion)
        if 1 <= seleccion <= len(canciones) * 3:
            return seleccion
        else:
            print("Entrada no válida. Por favor, ingrese un número válido.")
            seleccionar_cancion()
    else:
        print("Entrada no válida. Por favor, ingrese un número válido.")
        seleccionar_cancion()

def obtener_letra(artista, cancion):
    url = f"https://api.lyrics.ovh/v1/{artista}/{cancion}"
    respuesta = requests.get(url)
    if respuesta.status_code == 200:
        data = respuesta.json()
        if 'lyrics' in data:
            return data['lyrics']
    return "Lo siento, la letra de esta canción no está disponible."

def main():
    mostrar_menu()
    seleccion = seleccionar_cancion()
    index_cancion = (seleccion - 1) % 3
    index_artista = (seleccion - 1) // 3
    artista = list(canciones.keys())[index_artista]
    cancion = canciones[artista][index_cancion]
    letra = obtener_letra(artista, cancion)
    print(f"\n------ {cancion} by {artista} ------\n{letra}")
    if input("\nPresione * para elegir nuevamente: ") == "*":
        main()

if __name__ == "__main__":
    main()
````

## Análisis del Código

El script consta de las siguientes partes:

1. **Diccionario `canciones`:**
   - Contiene una lista de artistas como claves y una lista de sus canciones como valores.

2. **Función `mostrar_menu()`:**
   - Muestra un menú de canciones disponibles al usuario.

3. **Función `seleccionar_cancion()`:**
   - Permite al usuario seleccionar una canción ingresando su número correspondiente.
   - Verifica si la entrada del usuario es válida y solicita una nueva entrada si no lo es.

4. **Función `obtener_letra(artista, cancion)`:**
   - Utiliza la API `api.lyrics.ovh` para obtener la letra de la canción seleccionada.
   - Realiza una solicitud HTTP y procesa la respuesta para extraer la letra de la canción.

5. **Función `main()`:**
   - Llama a `mostrar_menu()` para mostrar las opciones de canciones disponibles.
   - Llama a `seleccionar_cancion()` para permitir al usuario seleccionar una canción.
   - Utiliza los índices de selección para obtener el nombre del artista y la canción seleccionados.
   - Llama a `obtener_letra()` para obtener la letra de la canción seleccionada.
   - Muestra la letra de la canción al usuario.
   - Si el usuario desea elegir otra canción, se le ofrece la opción de hacerlo.

6. **Bloque de Ejecución Principal:**
   - Verifica si el script se está ejecutando directamente y, si es así, llama a la función `main()`.

El código está diseñado para proporcionar una experiencia interactiva donde los usuarios pueden explorar letras de canciones de manera fácil y conveniente.

# Ejercicio #7 "Palíndromo"

En este ejercicio, desarrollaremos un programa en Python llamado "Palíndromo" que permitirá al usuario ingresar cinco palabras y determinar si alguna de ellas es un palíndromo. Un palíndromo es una palabra, frase, número u otra secuencia de caracteres que se lee igual hacia adelante y hacia atrás (ignorando espacios, puntuación y diferencia entre mayúsculas y minúsculas).

## Introducción al Programa

El programa solicitará al usuario que ingrese cinco palabras. Luego, verificará cada palabra para determinar si es un palíndromo. Si alguna de las palabras ingresadas es un palíndromo, el programa mostrará todas las palabras palíndromas encontradas. Si ninguna de las palabras es un palíndromo, el programa mostrará un mensaje indicando que ninguna de las palabras ingresadas es un palíndromo.

### Código Python 


````python
def es_palindromo(palabra):
    return palabra == palabra[::-1]

def main():
    palabras = []
    for i in range(5):
        palabra = input(f"Ingrese la palabra {i+1}: ").lower()
        palabras.append(palabra)

    palindromos = []
    for palabra in palabras:
        if es_palindromo(palabra):
            palindromos.append(palabra)

    if palindromos:
        print("Las siguientes palabras son palíndromos:")
        for palindromo in palindromos:
            print("-", palindromo)
    else:
        print("Ninguna de las palabras ingresadas es un palíndromo.")

if __name__ == "__main__":
    main()
````

## Análisis del Código

El script consta de dos funciones principales:

1. **Función `es_palindromo(palabra)`:**
   - Esta función toma una palabra como entrada y determina si es un palíndromo comparando la palabra original con su versión invertida utilizando la técnica de "slicing".
   - Retorna `True` si la palabra es un palíndromo y `False` en caso contrario.

2. **Función `main()`:**
   - Solicita al usuario que ingrese cinco palabras.
   - Utiliza un bucle `for` para iterar sobre cada palabra ingresada y determinar si es un palíndromo utilizando la función `es_palindromo()`.
   - Guarda las palabras palíndromas encontradas en una lista.
   - Si se encuentran palabras palíndromas, las imprime. De lo contrario, imprime un mensaje indicando que ninguna de las palabras ingresadas es un palíndromo.

El código proporciona una manera simple pero efectiva de determinar si las palabras ingresadas por el usuario son palíndromos. Es interactivo y ofrece retroalimentación clara sobre los resultados obtenidos.


# Ejercicio #8 "parImpar"

En este ejercicio, desarrollaremos un programa en Python llamado "parImpar" que permitirá al usuario ingresar números y determinar si son pares o impares. El programa solicitará al usuario que introduzca un número entre 1 y 1000, verificará si es par o impar y mostrará el resultado. Luego, preguntará al usuario si desea probar con otro número.

## Introducción al Programa

El programa solicitará al usuario que introduzca un número entero entre 1 y 1000. Luego, verificará si el número ingresado es par o impar utilizando operadores aritméticos y la operación de módulo (%). Después de determinar si el número es par o impar, el programa preguntará al usuario si desea probar con otro número. Si el usuario responde "Sí", el programa continuará solicitando números. Si el usuario responde "No" o cualquier otra entrada que no sea "Sí", el programa terminará.

### Código Python 

````python
def main():
    print("¡Bienvenido!")
    while True:
        try:
            numero = int(input("Por favor, introduce un número entre 1 y 1000: "))
            if 1 <= numero <= 1000:
                if numero % 2 == 0:
                    print("¡Ese número es par!")
                else:
                    print("¡Ese número es impar!")
                respuesta = input("¿Deseas probar con otro número? (Sí/No): ")
                if respuesta.lower() != 'si':
                    break
            else:
                print("Por favor, introduce un número válido entre 1 y 1000.")
        except ValueError:
            print("Por favor, introduce un número entero válido.")

if __name__ == "__main__":
    main()
````

## Análisis del Código

El script consta de una función principal `main()` que implementa la lógica del programa:

- Utiliza un bucle `while True` para solicitar continuamente al usuario que ingrese números.
- Utiliza un bloque `try-except` para manejar posibles errores de conversión de tipo si el usuario ingresa caracteres no numéricos.
- Verifica si el número ingresado está dentro del rango válido (entre 1 y 1000) y determina si es par o impar.
- Solicita al usuario si desea probar con otro número y, si la respuesta es negativa, rompe el bucle y termina el programa.

El código proporciona una manera interactiva y efectiva de determinar si los números ingresados por el usuario son pares o impares, ofreciendo una experiencia amigable para el usuario.

## Introducción al Juego

El juego comenzará mostrando al jugador las opciones disponibles (piedra, papel o tijeras). Luego, solicitará al jugador que ingrese su elección. Después de que el jugador haya hecho su elección, el programa generará una elección aleatoria para la computadora y determinará el resultado del juego. Si el jugador decide salir del juego, puede hacerlo ingresando la opción "4". En ese caso, el programa mostrará un mensaje de despedida y terminará.

### Código Python 

````python
import random

def jugar_piedra_papel_tijeras(jugador, computadora):
    if jugador == computadora:
        return "¡Es un empate!"
    elif (jugador == 1 and computadora == 3) or (jugador == 2 and computadora == 1) or (jugador == 3 and computadora == 2):
        return "¡Felicidades! ¡Ganaste!"
    else:
        return "¡Oh no! ¡La computadora ganó!"

def main():
    opciones = {1: "piedra", 2: "papel", 3: "tijeras"}
    while True:
        print("Elige una opción:")
        print("1) Piedra")
        print("2) Papel")
        print("3) Tijeras")
        print("4) Salir")
        jugador = input("Tu elección: ")
        
        if jugador == '4':
            print("¡Gracias por jugar! ¡Hasta luego!")
            break

        try:
            jugador = int(jugador)
            if jugador in opciones:
                computadora = random.randint(1, 3)
                resultado = jugar_piedra_papel_tijeras(jugador, computadora)
                print("Tú elegiste:", opciones[jugador])
                print("La computadora eligió:", opciones[computadora])
                print(resultado)
            else:
                print("Por favor, elige una opción válida.")
        except ValueError:
            print("Por favor, ingresa un número válido.")

if __name__ == "__main__":
    main()
````

## Análisis del Código

El script consta de una función principal `main()` que implementa la lógica del juego:

- Utiliza un bucle `while True` para permitir que el jugador juegue varias veces hasta que decida salir.
- Muestra las opciones disponibles al jugador y solicita al jugador que ingrese su elección.
- Verifica si la elección del jugador es válida (es decir, un número entre 1 y 4) y maneja posibles errores de entrada.
- Si el jugador decide salir, muestra un mensaje de despedida y termina el programa.
- Si el jugador elige una opción válida, genera una elección aleatoria para la computadora y determina el resultado del juego utilizando la función `jugar_piedra_papel_tijeras()`.
- Muestra las elecciones del jugador y la computadora, así como el resultado del juego.

El código proporciona una versión simplificada pero divertida del juego "Piedra, Papel o Tijeras", ofreciendo una experiencia interactiva para el jugador.

# Ejercicio #10 "propina"

En este ejercicio, desarrollaremos un programa en Python llamado "propina" que ayudará a calcular la propina recomendada para una cuenta de restaurante. El programa solicitará al usuario que ingrese el total de la cuenta y calculará tres opciones diferentes de propina: 18%, 20% y 25%. Luego, mostrará el monto de la propina y el total de la cuenta, incluyendo la propina, para cada una de las tres opciones.

## Introducción al Programa

El programa comenzará solicitando al usuario que ingrese el total de la cuenta del restaurante. Luego, utilizará esta información para calcular tres opciones diferentes de propina: 18%, 20% y 25%. Para cada opción de propina, el programa calculará el monto de la propina y el total de la cuenta, incluyendo la propina. Finalmente, mostrará estos montos al usuario.

### Codigo Python


````python
def calcular_propina(total_cuenta):
    propina_18 = total_cuenta * 0.18
    total_18 = total_cuenta + propina_18

    propina_20 = total_cuenta * 0.20
    total_20 = total_cuenta + propina_20

    propina_25 = total_cuenta * 0.25
    total_25 = total_cuenta + propina_25

    return propina_18, total_18, propina_20, total_20, propina_25, total_25

def main():
    total_cuenta = float(input("¿Cuál es el total de la cuenta para hoy, por favor? $"))
    propina_18, total_18, propina_20, total_20, propina_25, total_25 = calcular_propina(total_cuenta)

    print(f"La propina del 18% es de ${propina_18:.2f}, lo que hace un total de ${total_18:.2f}.")
    print(f"La propina del 20% es de ${propina_20:.2f}, lo que hace un total de ${total_20:.2f}.")
    print(f"La propina del 25% es de ${propina_25:.2f}, lo que hace un total de ${total_25:.2f}.")

if __name__ == "__main__":
    main()
````

## Análisis del Código

El script consta de dos funciones principales:

1. **Función `calcular_propina(total_cuenta)`:**
   - Esta función toma el total de la cuenta como entrada y calcula tres opciones diferentes de propina (18%, 20% y 25%), así como el total de la cuenta incluyendo cada propina.
   - Retorna los montos de la propina y los totales de la cuenta para cada opción.

2. **Función `main()`:**
   - Solicita al usuario que ingrese el total de la cuenta del restaurante.
   - Llama a la función `calcular_propina()` para calcular las opciones de propina y los totales de la cuenta.
   - Imprime los montos de la propina y los totales de la cuenta para cada opción en el formato adecuado.

El código proporciona una manera sencilla pero efectiva de calcular la propina recomendada para una cuenta de restaurante, ofreciendo al usuario varias opciones para elegir.

# Ejercicio #11 "WordCount"

En este ejercicio, desarrollaremos un programa en Python llamado "WordCount" que contará el número de palabras en una oración proporcionada por el usuario y en un archivo de texto especificado. El programa solicitará al usuario que ingrese una oración y calculará el número de palabras en esa oración. Luego, solicitará al usuario que ingrese el nombre de un archivo de texto y calculará el número de palabras en ese archivo.

## Introducción al Programa

El programa comenzará solicitando al usuario que ingrese una oración. Luego, utilizará la función `contar_palabras()` para contar el número de palabras en la oración y mostrará este número al usuario. Después, solicitará al usuario que ingrese el nombre de un archivo de texto. Si el archivo existe, el programa abrirá el archivo, leerá su contenido, contará el número de palabras en el contenido del archivo y mostrará este número al usuario. Si el archivo no existe, el programa mostrará un mensaje de error.

### Codigo Python


````python
def contar_palabras(oracion):
    palabras = oracion.split()
    numero_palabras = len(palabras)
    return numero_palabras

def main():
    oracion_usuario = input("¿Qué tienes en mente hoy? ")
    palabras_oracion_usuario = contar_palabras(oracion_usuario)
    print(f"Oh, ¡qué bueno! ¡Acabas de decirme lo que tienes en mente en {palabras_oracion_usuario} palabras!")

    nombre_archivo = input("Por favor, introduce el nombre del archivo: ")
    try:
        with open(nombre_archivo, 'r') as archivo:
            contenido = archivo.read()
            palabras_archivo = contar_palabras(contenido)
            print(f"El archivo '{nombre_archivo}' tiene un total de {palabras_archivo} palabras.")
    except FileNotFoundError:
        print("El archivo no se encontró. Por favor, asegúrate de que el nombre del archivo sea correcto.")

if __name__ == "__main__":
    main()
````

## Análisis del Código

El script consta de dos funciones principales:

1. **Función `contar_palabras(oracion)`:**
   - Esta función toma una oración como entrada, la divide en palabras utilizando el método `split()` y cuenta el número de palabras.
   - Retorna el número de palabras en la oración.

2. **Función `main()`:**
   - Solicita al usuario que ingrese una oración y llama a la función `contar_palabras()` para contar el número de palabras en esa oración.
   - Solicita al usuario que ingrese el nombre de un archivo de texto y trata de abrir el archivo.
   - Si el archivo se abre correctamente, lee su contenido, llama a la función `contar_palabras()` para contar el número de palabras en el contenido del archivo y muestra este número.
   - Si el archivo no se encuentra, muestra un mensaje de error.

El código proporciona una manera eficiente de contar el número de palabras en una oración y en un archivo de texto, ofreciendo una experiencia útil para el usuario.

# Ejercicio #2 "Adivina el Número"

Este notebook proporciona un análisis del juego "Adivina el Número" implementado en Python y documenta su funcionamiento.

## Código Python


````python
import random

def jugar_adivina_el_numero():
    """
    Esta función implementa el juego "Adivina el Número".
    El jugador debe adivinar un número secreto entre 1 y 50.
    """
    numero_secreto = random.randint(1, 50)
    intentos = 0
    
    while True:
        intento = input("Adivina el número secreto entre 1 y 50: ")

        try:
            intento = int(intento)
            if intento < 1 or intento > 50:
                print("Error: Por favor, elige un número dentro del rango correcto (1-50).")
                continue
            
            intentos += 1
            if intento < numero_secreto:
                print("Demasiado bajo. Intenta de nuevo.")
            elif intento > numero_secreto:
                print("Demasiado alto. Intenta de nuevo.")
            else:
                print(f"Felicidades, ¡adivinaste el número secreto {numero_secreto} en {intentos} intentos!")
                break

            continuar = input("¿Quieres seguir jugando? (sí/no): ").lower()
            if continuar != 'si':
                print("¡Gracias por jugar!")
                break
        except ValueError:
            print("Error: Por favor, ingresa un número válido.")

def main():
    print("Bienvenido al juego Adivina el Número!")
    jugar_adivina_el_numero()

if __name__ == "__main__":
    main()
````

## Análisis del Código

El juego "Adivina el Número" tiene las siguientes características:

- La función `jugar_adivina_el_numero()` genera un número secreto aleatorio entre 1 y 50 y solicita al usuario que adivine ese número.
- El jugador tiene un número limitado de intentos para adivinar el número secreto.
- Después de cada intento, se proporciona una pista para indicar si el número es demasiado bajo o demasiado alto.
- Una vez que el jugador adivina correctamente el número secreto, se muestra un mensaje de felicitaciones junto con el número de intentos realizados.
- Después de cada juego, se le pregunta al jugador si desea continuar jugando.
- La función `main()` inicia el juego y da la bienvenida al jugador.

El código está diseñado para manejar errores como la entrada de valores no válidos y garantizar que el jugador elija un número dentro del rango especificado. El juego proporciona una experiencia interactiva y divertida para el usuario.


# Ejercicio #3 "Biografía"

En este ejercicio, crearemos un programa en Python llamado "Biografía" que solicitará información personal al usuario, como nombre, fecha de nacimiento, dirección y metas personales. El programa validará cada entrada de información para asegurarse de que esté en el formato correcto antes de mostrar un resumen de la información ingresada.

El programa utilizará expresiones regulares para validar cada tipo de información introducida por el usuario, como el nombre, la fecha de nacimiento y la dirección. Si la información ingresada no cumple con los requisitos, se solicitará al usuario que la ingrese nuevamente.

Una vez que se haya ingresado toda la información válida, el programa imprimirá un resumen de la información personal del usuario.

## Código Python 



````python
import re

def validar_nombre(nombre):
    if re.match("^[A-Za-z ]+$", nombre):
        return True
    else:
        return False

def validar_fecha(fecha):
    if re.match("^[A-Za-z]+ [0-9]{1,2}, [0-9]{4}$", fecha):
        return True
    else:
        return False

def validar_direccion(direccion):
    if re.match("^[A-Za-z0-9, ]+$", direccion):
        return True
    else:
        return False

def solicitar_informacion():
    nombre = input("¿Cuál es tu nombre? ")
    while not validar_nombre(nombre):
        print("Por favor, ingresa un nombre válido.")
        nombre = input("¿Cuál es tu nombre? ")

    fecha_nacimiento = input("¿Cuál es tu fecha de nacimiento? (Ejemplo: Jun 1, 1954) ")
    while not validar_fecha(fecha_nacimiento):
        print("Por favor, ingresa una fecha de nacimiento válida en el formato correcto.")
        fecha_nacimiento = input("¿Cuál es tu fecha de nacimiento? (Ejemplo: Jan 1, 1954) ")

    direccion = input("¿Cuál es tu dirección? ")
    while not validar_direccion(direccion):
        print("Por favor, ingresa una dirección válida.")
        direccion = input("¿Cuál es tu dirección? ")

    metas_personales = input("¿Cuáles son tus metas personales? ")

    return nombre, fecha_nacimiento, direccion, metas_personales

def imprimir_resumen(nombre, fecha_nacimiento, direccion, metas_personales):
    print("\nResumen de tu información personal:")
    print("- Nombre:", nombre)
    print("- Fecha de nacimiento:", fecha_nacimiento)
    print("- Direccion:", direccion)
    print("- Metas:", metas_personales)

def main():
    nombre, fecha_nacimiento, direccion, metas_personales = solicitar_informacion()
    imprimir_resumen(nombre, fecha_nacimiento, direccion, metas_personales)

if __name__ == "__main__":
    main()
````

# Análisis del Código

El script consta de las siguientes partes:

1. **Funciones de Validación:**
   - `validar_nombre(nombre)`: Utiliza una expresión regular para validar que el nombre ingresado contenga solo letras y espacios.
   - `validar_fecha(fecha)`: Utiliza una expresión regular para validar que la fecha de nacimiento ingresada esté en el formato correcto (por ejemplo, "Jun 1, 1954").
   - `validar_direccion(direccion)`: Utiliza una expresión regular para validar que la dirección ingresada contenga solo letras, números, comas y espacios.

2. **Función `solicitar_informacion()`:**
   - Solicita al usuario que ingrese su nombre, fecha de nacimiento y dirección.
   - Utiliza las funciones de validación para asegurarse de que cada entrada sea válida.
   - Devuelve la información ingresada por el usuario.

3. **Función `imprimir_resumen()`:**
   - Toma los datos personales ingresados por el usuario y los imprime en un formato de resumen.

4. **Función `main()`:**
   - Llama a la función `solicitar_informacion()` para obtener la información del usuario.
   - Llama a la función `imprimir_resumen()` para mostrar un resumen de la información ingresada.

5. **Bloque de Ejecución Principal:**
   - Verifica si el script se está ejecutando directamente y, si es así, llama a la función `main()`.
